```python
from transformers import pipeline
import json
```

```python
file_path = '/content/Data (1).json'
```

```python
with open(file_path, 'r') as file:
    patent = json.load(file)
```

```python
patent
```

```python
for patents in patent:
    input_text = patents.get('claims','')
    print(input_text)
```

```python
for patents in patent:
    input_text = patents.get('abstract','')
    print(input_text)
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
abstract_summaries = []
claims_summaries = []
```

```python
for patents in patent:
    input_text = patents.get('abstract','')
    if input_text:
      abstract_summary = summarizer(input_text, max_length=39, min_length=30, do_sample=False)[0]['summary_text']
    else:
      abstract_summary = ""
    abstract_summaries.append(abstract_summary)
```

```python
abstract_summaries
```

```python
for patents in patent:
    input_text = patents.get('claims','')
    if input_text:
      claims_summary = summarizer(input_text, max_length=13, min_length=10, do_sample=False)[0]['summary_text']
    else:
      claims_summary = ""
    claims_summaries.append(claims_summary)
```

```python
claims_summaries
```

```python
output_json = {
    "claims": claims_summaries,
    "abstract": abstract_summaries
}
```
